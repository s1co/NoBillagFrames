### Teil der Daten kodieren

```python
# Mittels train_test_split() randomisiert Tweets ausw√§hlen, die Manuell kodiert werden sollen
from sklearn.model_selection import train_test_split
data_labeled, data_unlabeled  = train_test_split(data_raw, test_size=0.99)
data_labeled.shape
```

```python
# Tweets Manuell kodieren
def label_tweets(text):
    label_desc = "Codes: (0) Kann nicht zugordnet werden / (1) Pressefreiheit / (2) Kommerzialisierung / (3) Mediale Grundversorgung / (4) Finanzen Haushalte / (5) SRG / (6) Libertarismus / (7) Pro / (8) Contra \nFolgenden Tweet zu ordnen:\n'"
    code = input(label_desc + text + "':\n")
    return code
data_labeled['Label'] = data_labeled['Text'].apply(label_tweets)
```
